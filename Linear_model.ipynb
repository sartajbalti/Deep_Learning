b = bais

- LM can be used with multiple inputs to get multiple outputs. to find two linear relationships.

```python
# Simple Linear Regression 
# Import the relevent libraries
import numpy as np 
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# Generate a new data to train on the model
obeser = 1000
xs =np.random.uniform(-10,10,obeser)
zs = np.random.uniform(-10,10,obeser)
inputs = np.column_stack((xs,zs))

print(inputs.shape)
```

```python
# Create the traget we will aim at 
noise = np.random.uniform(-1,1,(obeser))
targets = 3*xs + 2*zs +5+ noise
```

```python
print(targets.shape)
```

```python
# In order to use the 3D plot, the objects should have a certain shape, so we reshape the targets.
# The proper method to use is reshape and takes as arguments the dimensions in which we want to fit the object.
#targets = targets.reshape(obeser,)

# Plotting according to the conventional matplotlib.pyplot syntax

# Declare the figure
fig = plt.figure()

# A method allowing us to create the 3D plot
ax = fig.add_subplot(111, projection='3d')

# Choose the axes.
ax.plot(xs, zs, targets)

# Set labels
ax.set_xlabel('xs')
ax.set_ylabel('zs')
ax.set_zlabel('Targets')

# You can fiddle with the azim parameter to plot the data from different angles. Just change the value of azim=100
# to azim = 0 ; azim = 200, or whatever. Check and see what happens.
ax.view_init(azim=100)

# So far we were just describing the plot. This method actually shows the plot. 
plt.show()

# We reshape the targets back to the shape that they were in before plotting.
# This reshaping is a side-effect of the 3D plot. Sorry for that.
targets = targets.reshape(obeser,1)
```

```python
# Initislize variables
init_range = 0.1
weights = np.random.uniform(-init_range,init_range,(2,1))
bais = np.random.uniform(-init_range,init_range,(1,1))
print(weights)
print(bais)
```

```python
# initialize the learning rate
lr = 0.01
```

```python
# Train the model
for i in range(1000):
    outputs = np.dot(inputs,weights) + bais
    deltas = outputs - targets
    loss = np.sum(deltas**2) / 2/obeser
    print(loss)
    
    deltas_scale = deltas / obeser
    weights = weights - lr * np.dot(inputs.T,deltas_scale)
    bais = bais - lr * np.sum(deltas_scale)
    
```

```python
print(weights)
print(bais)
```

```python
plt.plot(outputs,targets,'o')
plt.xlabel('outputs')
plt.ylabel('targets')
```
